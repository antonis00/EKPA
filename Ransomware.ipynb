```python


import pandas as pd
import numpy as np
import ppscore as pps
import seaborn as sns

```

```python
dataset=pd.read_csv("https://media.githubusercontent.com/media/kdemertzis/EKPA/main/Data/Ransomware.csv",sep='|')
numeric_cols = dataset.select_dtypes(include=[np.number])
```

```python
numeric_cols
```

```python
from matplotlib import pyplot as plt
numeric_cols['MajorLinkerVersion'].plot(kind='hist', bins=20, title='MajorLinkerVersion')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
from matplotlib import pyplot as plt
numeric_cols.plot(kind='scatter', x='MajorLinkerVersion', y='MinorLinkerVersion', s=32, alpha=.8)
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
from matplotlib import pyplot as plt
numeric_cols['MajorLinkerVersion'].plot(kind='line', figsize=(8, 4), title='MajorLinkerVersion')
plt.gca().spines[['top', 'right']].set_visible(False)
```

```python
matrix_df = pps.matrix(numeric_cols)[['x', 'y', 'ppscore']].pivot(columns='x', index='y', values='ppscore')
```

```python
matrix_df
```

```python
plt.figure(figsize=(40, 20))  # You can adjust these numbers to suit your needs.
sns.heatmap(matrix_df, vmin=0, vmax=1, cmap="Blues", linewidths=1.5, annot=True)
```
